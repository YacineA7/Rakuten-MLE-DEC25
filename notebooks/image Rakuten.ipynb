````python
def traitement_rgb(chemin_image):
    # Chargement de l'image
    image_bgr = cv2.imread(chemin_image)
    
    if image_bgr is None:
        print(f"Impossible de charger l'image : {chemin_image}")
        return
    
    # Conversion BGR (OpenCV) → RGB
    image_rgb = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2RGB)
    
    # Séparation des canaux R, G, B
    canal_r = image_rgb[:, :, 0]
    canal_g = image_rgb[:, :, 1]
    canal_b = image_rgb[:, :, 2]
    
    # Statistiques par canal
    for nom, canal in zip(['Rouge', 'Vert', 'Bleu'], [canal_r, canal_g, canal_b]):
        print(f"Canal {nom} — Min: {canal.min()}, Max: {canal.max()}, Moyenne: {canal.mean():.2f}")
    
    # Affichage de l'image et de ses canaux
    fig, axes = plt.subplots(1, 4, figsize=(16, 4))
    
    axes[0].imshow(image_rgb)
    axes[0].set_title("Image originale")
    axes[0].axis('off')
    
    for ax, canal, titre, cmap in zip(
        axes[1:],
        [canal_r, canal_g, canal_b],
        ['Canal Rouge', 'Canal Vert', 'Canal Bleu'],
        ['Reds', 'Greens', 'Blues']
    ):
        ax.imshow(canal, cmap=cmap)
        ax.set_title(titre)
        ax.axis('off')
    
    plt.tight_layout()
    plt.show()
    
    return image_rgb, canal_r, canal_g, canal_b


# --- Utilisation ---
chemin = "votre_image.jpg"  # ← remplacez par votre chemin
traitement_rgb(chemin)
````

````python
def pixeliser(image_rgb, taille_bloc=10):
    h, w, _ = image_rgb.shape
    image_pixel = image_rgb.copy()
    
    for y in range(0, h, taille_bloc):
        for x in range(0, w, taille_bloc):
            # On prend la couleur moyenne du bloc
            bloc = image_rgb[y:y+taille_bloc, x:x+taille_bloc]
            couleur_moyenne = bloc.mean(axis=(0, 1)).astype(int)
            # On remplace tout le bloc par cette couleur
            image_pixel[y:y+taille_bloc, x:x+taille_bloc] = couleur_moyenne
    
    return image_pixel


# Plus `taille_bloc` est grand, plus l'effet de pixélisation est prononcé.

# **Résumé visuel**
# ```
# Taille bloc = 1  →  Image normale
# Taille bloc = 5  →  Légèrement pixélisée
# Taille bloc = 20 →  Très pixélisée
````
